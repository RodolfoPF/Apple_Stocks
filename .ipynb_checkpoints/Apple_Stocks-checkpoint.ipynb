# Predição do preço de ações da Apple.

```python
# Importando as bibliotecas.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential # Classe utilizada para criar o modelo sequencial utilizando o keras
from keras.layers import Dense # Classe utilizada para criar as camadas que são completamente conectadas
from keras.layers import LSTM # Classe para a rede recorrente utilizando long short term memory
from keras.layers import Dropout # Classe utilizada para a camada de dropout (utilizada para evitar overfiting)
%matplotlib inline
```

```python
# Carregando os dados

data = pd.read_csv("AAPL.csv")
```

```python
# Separando os dados para servirem de entrada para o nosso modelo.

entrada = data.iloc[:,1:2].values # Estou selecionando do dataframe somente a coluna referente aos preços de abertura.
entrada[:5] # Visualizando as 5 valores
```

```python
# Normalizando dados.

scaler = MinMaxScaler(feature_range=(0, 1))
# Criamos o objeto que será utilizado para realizar a normalizaçãodos dados.
# O feature_range define o intervalo de escala ao qual os dados serão normalizados.

dados_normalizados = scaler.fit_transform(entrada)
```

```python
dados_normalizados[:5]
```

```python
# Como as redes recorrentes utilizam dados no tempo T e valores passados (T - n), a entrada da rede deve conter os valores
# presentes e os (T - n). Assim, é necessário realizar uma modificação nos dados.

features_set = []
labels = []
for i in range(60, 1259):
    features_set.append(dados_normalizados[i-60:i, 0])
    labels.append(dados_normalizados[i, 0])
```

```python
# Transformando os dados em um array para serem utilizados como entrada.

features_set, labels = np.array(features_set), np.array(labels)
```

```python
# Conferindo as dimensões dos dados.

print(features_set.shape)
print(labels.shape)
```

```python
# Transformando os dados para o foramto aceito pelas redes recorrentes do Keras
# Formato em 3D
# (a,b,c) -> a = Número de linhas do dataset
#            b = Número de steps (entradas) da rede
#            c = Número de saídas.
# Método da biblioteca Numpy que é utilizado para converter os dados de entrada (1199, 60) em (1199,60,1)

features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))
print(features_set)
```

```python
model = Sequential() # Objeto para a criação do modelo Keras Sequential.
model.add(LSTM(units=50, return_sequences=True, input_shape=(features_set.shape[1], 1))) # Cria a camada de entrada

# Como pode ser visto, ela é adicionada como uma pilha, cada camada é adicionada com o método add na camada de entrada. é
# necessário definir o tamanho do vetor de entrada (input_shape), units = 50 indica que na camada de entrada devem existir
# 50 neurônios, return_sequences = True indicaque devem se adicionadas novas camadas.
```

```python
# Adição da camada de dropout.

model.add(Dropout(0.2)) # O valor de 0.2 indica que 20% dos neurônios dessa camada serão perdidos para cada interação
```

```python
# Adicionando mais camadas ao modelo.

model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50)) # Como essa é a última camada de LSTM utilizada, a variável return_sequences=False
model.add(Dropout(0.2))
```

```python
# Adicionando a camada de saída com apenas 1 neurônio, pois vamos realizar a previsão de apenas uma variável
# (Previsão do valor de abertura).

model.add(Dense(units=1))
```

```python
# Visualizando as configurações do nosso modelo.

model.summary()
```

```python
# Definição do tipo de função de perda a ser utilizada e do tipo do otimizador, o otimizador é utilizado para minimizar a 
# função perda, a função perda indica como deve ser calculado o erro do modelo (valor real - valor previsto).

model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
# Treinando o modelo

model.fit(features_set, labels, epochs=100, batch_size=32)
# Valores de entrada
# Valores de Saída
# Número de épocas para o treinamento (vezes em que vamos realizar interações durante o treinamento)
# Batch_size = quantidade de dados utilizados por vez para realizar o treinamento
```

```python
# Importando os dados de teste.

data_test = pd.read_csv("AAPL_previsao.csv")
```

```python
data_test.head()
```

```python
data_test_ajusted = data_test.iloc[:, 1:2].values
```

```python
data_test_ajusted = pd.concat((data["Open"], data_test["Open"]), axis=0)
```

```python
data_test_ajusted.head()
```

```python
test_inputs = data_test_ajusted[len(data_test_ajusted) - len(data_test) - 60:].values
```

```python
# Normalizando os dados para teste, como fizemos com os dados de treinamento

test_inputs = test_inputs.reshape(-1,1)
test_inputs = scaler.transform(test_inputs)
```

```python
# Preparação dos 60 dados a serem utilizados

test_features = []
for i in range(60, 80):
    test_features.append(test_inputs[i-60:i, 0])
```

```python
# Preparando os dados como entrada para o modelo de previsão

test_features = np.array(test_features)
test_features = np.reshape(test_features, (test_features.shape[0], test_features.shape[1], 1))
```

```python
# Previsão utilizando o modelo gerado

predict = model.predict(test_features)
```

```python
# Invertendo a transformação (normalização) dos dados de previsão

predict = scaler.inverse_transform(predict)
```

```python
# Plotando resultado da previsão e o real

plt.figure(figsize=(12,8))
plt.plot(data_test.Open, color="blue", label="Preço real das ações da Apple")
plt.plot(predict, color="red", label="Previsão do preço das ações da Apple")
plt.title("Previsão do preço de abertura das ações da Apple")
plt.xlabel("Data")
plt.ylabel("Preço")
plt.legend()
```
